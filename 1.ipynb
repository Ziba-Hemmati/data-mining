```python
# وارد کردن کتابخانه‌های مورد نیاز برای پروژه
# pandas: برای پردازش و تجزیه‌وتحلیل داده‌ها در قالب جدول
import pandas as pd

# matplotlib: برای رسم نمودارهای مختلف و نمایش داده‌ها به صورت گرافیکی
import matplotlib.pyplot as plt

# seaborn: matplotlib برای ایجاد نمودارهای آماری پیشرفته، بر اساس 
import seaborn as sns
```

```python
# CSV بارگذاری داده‌ها از فایل 
data = pd.read_csv('D:/Programming/data-set/ab_testing.csv')
```

```python
# نمایش اطلاعات کلی از داده‌ها (تعداد ردیف‌ها، ستون‌ها، نوع داده‌ها و مقادیر غیر خالی)
data.info()

# نمایش نخستین 5 ردیف از داده‌ها برای بررسی سریع
data.head()
```

```python
print(data.isnull().sum())  # بررسی مقادیر خالی
print(data.duplicated().sum())  # بررسی داده‌های تکراری
```

```python
# رسم نمودار پراکندگی برای نمایش رابطه بین تعداد بازدید صفحات و زمان سپری‌شده
plt.scatter(data['Page Views'], data['Time Spent'])

# برچسب‌گذاری محور افقی (Page Views)
plt.xlabel('Page Views')

# برچسب‌گذاری محور عمودی (Time Spent به ثانیه)
plt.ylabel('Time Spent (seconds)')

# عنوان نمودار
plt.title('Scatter Plot: Page Views vs Time Spent')

# نمایش نمودار
plt.show()
```

```python
# رسم هیستوگرام زمان صرف‌شده با ۲۰ سطل
data['Time Spent'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Time Spent')
plt.xlabel('Time Spent (seconds)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Group', y='Time Spent', data=data)
plt.title('Box Plot: Time Spent by Group')
plt.show()
```

```python
group_counts = data['Group'].value_counts()
group_counts.plot(kind='bar', color=['orange', 'blue'], edgecolor='black')
plt.title('Bar Chart: User Counts by Group')
plt.xlabel('Group')
plt.ylabel('User Count')
plt.show()
```

```python
device_counts = data['Device'].value_counts()
device_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue'], startangle=90)
plt.title('Device Distribution')
plt.ylabel('')
plt.show()
```
